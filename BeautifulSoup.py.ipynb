```python
from bs4 import BeautifulSoup
import urllib
```

```python
#Getting all the links
handle=urllib.urlopen("https://www.nytimes.com/2017/11/27/us/politics/senate-tax-bills-potential-hurdle-republicans.html")
read=handle.read()

soup=BeautifulSoup(read,"html.parser")

para=soup('a')

for lines in para:
    if len(lines.contents)>=1:
        print lines.contents[0]
```

```python
#Getting all the paragraphs
handle=urllib.urlopen("https://www.nytimes.com/2017/11/27/us/politics/senate-tax-bills-potential-hurdle-republicans.html")
read=handle.read()

soup=BeautifulSoup(read,"html.parser")

para=soup('p')

for lines in para:
    if len(lines.contents)>=1:
        print lines.contents[0]
```
