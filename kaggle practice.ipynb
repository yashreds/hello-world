```python
data=pd.read_csv("suicide.csv",sep=",",date_parser="Year")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.head()
```

```python
pd.crosstab(index=data['Gender'],columns=data['Age_group'])
```

```python
data.groupby('State').sum()['Total'].plot("bar",figsize=(13,7),title='state wise report')
plt.show()
```

```python
data.groupby('Year').sum()['Total'].plot("bar",title='Year wise report')
plt.show()
```

```python
data.groupby('Type_code').sum()['Total'].plot('bar',title='Type_code wise report')
plt.show()
```

```python
data.groupby('Type').sum()['Total'].plot('bar',title='Type wise')
plt.show()
```

```python
data.groupby('Gender').sum()['Total'].plot('bar',title='Gender wise report')
plt.show()
```

```python
data[data['Age_group'] != '0-100+'].groupby('Age_group').sum()['Total'].plot('bar',title='Age group wise report')
plt.show()
```

```python
sb.factorplot(x='Type_code',y='Total',hue='Gender',data=data)
plt.show()
```

```python
sb.factorplot(x='Type',y='Total',kind='bar', data=data, estimator=sum ,size=6, aspect=2)
plt.show()
```

```python
sb.factorplot(x='Type',y='Total',hue='Gender',data=data,kind='bar',size=10)
plt.show()
```
